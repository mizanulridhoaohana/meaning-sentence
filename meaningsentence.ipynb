```python
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize

def is_sentence_meaningful(sentence):
    # Tokenisasi kalimat
    words = word_tokenize(sentence)

    # Menyimpan sinonim dari setiap kata dalam kalimat
    synonyms = set()
    for word in words:
        for syn in wordnet.synsets(word):
            for lemma in syn.lemmas():
                synonyms.add(lemma.name())

    # Memeriksa apakah ada sinonim yang ditemukan
    meaningful_synonyms = set()
    for word in words:
        for syn in wordnet.synsets(word):
            for lemma in syn.lemmas():
                meaningful_synonyms.add(lemma.name())

    return len(meaningful_synonyms) > 0

# Contoh penggunaan
kalimat = "aku mencintaimu"
hasil = is_sentence_meaningful(kalimat)

if hasil:
    print("Kalimat memiliki makna.")
else:
    print("Kalimat tidak memiliki makna.")
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import joblib

def train_and_save_model():
    # Load IndoBERT model
    model = SentenceTransformer('indobenchmark/indobert-base-p1')

    # Load a simple logistic regression model for binary classification
    clf = make_pipeline(StandardScaler(), LogisticRegression())

    # Load new training data from "KBBI_web.txt"
    with open("KBBI_web_id.txt", "r", encoding="utf-8") as file:
        training_data = [line.strip().split(', ') for line in file]

    sentences, labels = zip(*training_data)

    # Encode training sentences
    encoded_train_sentences = model.encode(sentences)

    # Fit the logistic regression model
    clf.fit(encoded_train_sentences, labels)

    # Save the model
    joblib.dump(clf, 'logical_model.joblib')

# Train and save the model
train_and_save_model()
```

```python
import joblib
from sentence_transformers import SentenceTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
def load_and_predict_model(sentence):
    # Load the model
    clf = joblib.load('./model_2.joblib')

    # Load IndoBERT model
    model = SentenceTransformer('indobenchmark/indobert-base-p1')

    # Encode the input sentence
    encoded_sentence = model.encode(sentence)

    # Predict the probability of being logical and meaningful
    prob_logical_and_meaningful = clf.predict_proba([encoded_sentence])[:, 1]

    print(f"Persentase Kemiripan: {prob_logical_and_meaningful[0]*100:.2f}%")

    return prob_logical_and_meaningful[0] > 0.5
```
